```python
import numpy as np
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
x, y = make_regression(100, 1, noise=5, bias=5)
x += 3
y = y**2
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(x, y)
plt.xlabel("features")
plt.ylabel("labels")
plt.title("Test Regression Data")
plt.show()
```

```python
from linear.regression import LinearRegressor, PolynomialRegressor
from kernel.regression import KernelRegressor
from neighbors.regression import KNNRegressor
```

```python
LR = LinearRegressor()
```

```python
LR.fit(x, y)
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(x, y)
plt.xlabel("features")
plt.ylabel("labels")
plt.title("Test Regression Data")
plt.plot(x, LR.predict(x), 'r')
plt.legend(["Actual", "Predicted"])
plt.show()
```

```python
PR = PolynomialRegressor(2)
```

```python
PR.fit(x, y)
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(x, y)
plt.xlabel("features")
plt.ylabel("labels")
plt.title("Test Regression Data")
plt.scatter(x, PR.predict(x))
plt.legend(["Actual", "Predicted"])
plt.show()
```

```python
NR = KNNRegressor(k=3)
```

```python
NR.fit(x, y)
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(x, y)
plt.xlabel("features")
plt.ylabel("labels")
plt.title("Test Regression Data")
plt.scatter(x, NR.predict(x))
plt.legend(["Actual", "Predicted"])
plt.show()
```

```python
KR1 = KernelRegressor(kernel='linear')
KR2 = KernelRegressor(kernel='poly')
KR3 = KernelRegressor(kernel='rbf')
```

```python
KR1.fit(x, y)
KR2.fit(x, y, bias=1, degree=2)
KR3.fit(x, y, sigma=0.5)
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(x, y)
plt.xlabel("features")
plt.ylabel("labels")
plt.title("Test Regression Data")
plt.scatter(x, KR1.predict(x))
plt.scatter(x, KR2.predict(x, bias=1, degree=2))
plt.scatter(x, KR3.predict(x, sigma=0.5))
plt.legend(["Actual", "Linear Kernel", "Polynomial Kernel", "Gaussian Kernel"])
plt.show()
```
